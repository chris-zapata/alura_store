#1. Análisis de facturación



```python
venta = tienda.Precio.sum()
venta2 = tienda2.Precio.sum()
venta3 = tienda3.Precio.sum()
venta4 = tienda4.Precio.sum()

print(f'La venta de la tienda 1 es de ${venta:,.0f}')
print(f'La venta de la tienda 2 es de ${venta2:,.0f}')
print(f'La venta de la tienda 3 es de ${venta3:,.0f}')
print(f'La venta de la tienda 4 es de ${venta4:,.0f}')
```

```python
import matplotlib.pyplot as plt

ventas_tiendas = [venta, venta2, venta3, venta4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

#Creamos gráfico para comparar las ventas de cada tienda
plt.bar(tiendas, ventas_tiendas, label='Ventas por tienda', color='lightblue')
plt.title('Total de ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas')
plt.show()
```

```python
plt.plot(tiendas, ventas_tiendas, label='Ventas por tienda', color='lightgray')
plt.title('Total de ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas')
plt.show()
```

```python
plt.scatter(tiendas, ventas_tiendas, label='Ventas por tienda', color='lightgray')
plt.title('Total de ventas por tienda')
plt.xlabel('Tienda')
plt.ylabel('Ventas')
plt.show()
```

La tienda que más vende es la tienda 1, mientras que la tienda 4 es la que cuenta con la menor facturación.

# 3. Calificación promedio de la tienda


```python
reputacion1 = tienda.Calificación.mean()
reputacion2 = tienda2.Calificación.mean()
reputacion3 = tienda3.Calificación.mean()
reputacion4 = tienda4.Calificación.mean()

print(f'La reputacion promedio de la tienda 1 es de {reputacion1:.2f}')
print(f'La reputacion promedio de la tienda 2 es de {reputacion2:.2f}')
print(f'La reputacion promedio de la tienda 3 es de {reputacion3:.2f}')
print(f'La reputacion promedio de la tienda 4 es de {reputacion4:.2f}')
```

```python
reputaciones = [reputacion1, reputacion2, reputacion3, reputacion4]
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
```

```python
fig, ax = plt.subplots(figsize=(8,4)) # figsize opcional
ax.barh(tiendas, reputaciones, color='lightblue', label='Reputación promedio')
for i, v in enumerate(reputaciones):
    texto_valor = f'{v:.3f}'
    ax.text(v,i,texto_valor,color='darkblue',va='center',fontsize=10)
ax.set_title('Reputación promedio por tienda')
ax.set_xlabel('Reputación promedio')
ax.set_ylabel('Tienda')
ax.legend()
plt.tight_layout()
plt.show()
```

En términos de reputación, la mejor tienda es la tienda 3, mientras que la tienda 1 es la con peor reputación.

# 4. Productos más y menos vendidos

```python
def detalle_productos(df, nombre):
  print(f'Los 3 productos más vendidos de la {nombre} son:')
  print(df['Producto'].value_counts().sort_values(ascending=False).head(3))
  print(f'Los 3 productos menos vendidos de la {nombre} son:')
  print(df['Producto'].value_counts().sort_values(ascending=False).tail(3))
  return df['Producto'].value_counts().sort_values(ascending=False).head(3), df['Producto'].value_counts().sort_values(ascending=False).tail(3)

detalle1_mas, detalle1_menos = detalle_productos(tienda, 'Tienda 1')
detalle2_mas, detalle2_menos = detalle_productos(tienda2, 'Tienda 2')
detalle3_mas, detalle3_menos = detalle_productos(tienda3, 'Tienda 3')
detalle4_mas, detalle4_menos = detalle_productos(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
def promedio_envio(df, nombre):
  print(f'El envío promedio de la {nombre} es de ${df["Costo de envío"].mean():,.2f}')
  return df["Costo de envío"].mean()

envio1 = promedio_envio(tienda, 'Tienda 1')
envio2 = promedio_envio(tienda2, 'Tienda 2')
envio3 = promedio_envio(tienda3, 'Tienda 3')
envio4 = promedio_envio(tienda4, 'Tienda 4')

envios = [envio1, envio2, envio3, envio4]
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(tiendas, envios, color='lightblue', label='Envío promedio')
for i, v in enumerate(envios):
    texto_valor = f'${v:,.1f}'
    ax.text(i,               # Posición X: en el centro horizontal de la barra
            v + 1,         # Posición Y: un poco por encima del final de la barra
            texto_valor,          # Texto a mostrar: el valor del envío
            color='black',
            ha='center',     # Alineación horizontal: centrado
            fontsize=10)
ax.set_title('Envío promedio por tienda')
ax.set_xlabel('Tienda')
ax.set_ylabel('Envío promedio')
ax.legend()
plt.tight_layout()
plt.show()
```

La tienda con el menor costo de envío promedio es la tienda 4, mientras que la tienda 1 posee el envío promedio más costoso

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import folium
from folium.plugins import HeatMap
from sklearn.cluster import DBSCAN


# unir los df en uno solo, además de agregarle la columna de tienda
t1 = tienda.assign(tienda='Tienda 1')
t2 = tienda2.assign(tienda='Tienda 2')
t3 = tienda3.assign(tienda='Tienda 3')
t4 = tienda4.assign(tienda='Tienda 4')

tiendas = pd.concat([t1, t2, t3, t4])
tiendas.sample(5)
```

```python
# Asegurar que las columnas críticas son numéricas
tiendas['lat'] = pd.to_numeric(tiendas['lat'], errors='coerce')
tiendas['lon'] = pd.to_numeric(tiendas['lon'], errors='coerce')
tiendas['Precio'] = pd.to_numeric(tiendas['Precio'], errors='coerce')

# Convertir fecha
tiendas['Fecha de Compra'] = pd.to_datetime(tiendas['Fecha de Compra'], errors='coerce')

# Eliminar filas sin coordenadas o sin precio
tiendas = tiendas.dropna(subset=['lat', 'lon', 'Precio'])

# Filtrar coordenadas inválidas
tiendas = tiendas[
    tiendas['lat'].between(-90, 90) &
    tiendas['lon'].between(-180, 180)
]
```

```python
plt.figure(figsize=(10,8))

tiendas_unicas = tiendas['tienda'].unique()
colors = plt.cm.tab10.colors  # 10 colores distintos

for i, t in enumerate(tiendas_unicas):
    sub = tiendas[tiendas['tienda'] == t]

    # Tamaño del punto proporcional al precio (solo visual, no análisis)
    sizes = np.clip(sub['Precio'] / sub['Precio'].max() * 100, 20, 200)

    plt.scatter(
        sub['lon'], sub['lat'],
        s=sizes,
        alpha=0.5,
        label=t,
        edgecolors='k',
        linewidth=0.3,
        color=colors[i % len(colors)]
    )

plt.xlabel("Longitud")
plt.ylabel("Latitud")
plt.title("Distribución geográfica de ventas por tienda")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Centro del mapa
center = [tiendas['lat'].mean(), tiendas['lon'].mean()]

m = folium.Map(location=center, zoom_start=11)

# Heatmap requiere formato:
# [lat, lon, peso]
heat_data = tiendas[['lat', 'lon', 'Precio']].values.tolist()

HeatMap(
    [[row[0], row[1], row[2]] for row in heat_data],
    radius=12,
    blur=15,
    max_zoom=1
).add_to(m)

m.save("heatmap_ventas.html")
print("Heatmap guardado como 'heatmap_ventas.html'. Ábrelo en tu navegador.")
```

```python
m2 = folium.Map(location=center, zoom_start=11)

for t in tiendas_unicas:
    sub = tiendas[tiendas['tienda'] == t]

    for _, row in sub.iterrows():
        folium.CircleMarker(
            location=[row['lat'], row['lon']],
            radius=3,
            popup=f"Tienda: {t} – Precio: ${row['Precio']:.1f}",
            tooltip=t,
        ).add_to(m2)

m2.save("mapa_por_tienda.html")
print("Mapa por tienda guardado como 'mapa_por_tienda.html'.")
```
